```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import interp1d
from scipy.interpolate import splrep, splev
```

```python
file="lab1kore.rad"
infile=open(file)
for line in infile.readlines():
    print(line)
```

```python
spectra = []
frequency = []
velocity = []
counter = 0
az = []
az1=[]

file="lab1kore.rad"
infile=open(file)

for line in infile.readlines():
    if ('*' not in line[0]):
        print(line)
        line_list = line.strip().split()
        str_spec = line_list[-64:]
        az = float(line_list[3])
        print(az)
        specific_spec = []
        for item in str_spec:
            specific_spec.append(float(item))
        plt.plot(specific_spec)
        plt.ylabel('Brightness Temperature(K)')
        plt.xlabel('Bin number')
        plt.show()
    #plt.savefig(’spectra.png’)
        spectra.append(specific_spec)
        az1.append(float(az))
infile.close()
```

```python
integrated_spectra = []

# you start with spectra[5] is because the ones before dont have an offset value
for spec in spectra[5:]:
    avg_int_spec = np.sum(spec)/len(spec)
    integrated_spectra.append(avg_int_spec)
    
# storing it in a numpy array
integrated_spectra = np.array(integrated_spectra)
```

```python
offset_to_match = np.array([-24., -21., -18., -15., -12., -9., -6., -3., 0., 3., 6., 9., 12., 15., 18., 21., 24.])
# multiplying with 0.84339 to correct the offsets so that the median of the beam pattern aligns at the center
offset_to_match_corr = offset_to_match * 0.84339
len(offset_to_match_corr)
```

```python
mean_intensity = []
std_intensity = []

for i, offset in enumerate(offset_to_match):
    p = np.argwhere(az1[5:]== offset).flatten()
    integrated_intensity_spec = []
    for i in p:
        integrated_intensity_spec.append(integrated_spectra[i])
        
    # average values
    mean_val = np.mean(integrated_intensity_spec)
    
    # standard deviations
    std_val = np.std(integrated_intensity_spec)
    mean_intensity.append(mean_val)
    std_intensity.append(std_val)
    
print(mean_intensity)
print(std_intensity)
```

```python
plt.figure(figsize=(10, 7))
plt.errorbar(
    offset_to_match_corr, 
    mean_intensity, 
    yerr=std_intensity, 
    ecolor='red',
    elinewidth=10,
)
plt.title("Average Integrated Intensity vs. Angular offsets")
plt.xlabel("Angular offsets (in degrees)")
plt.ylabel("Average Integrated Intensity")
plt.show()
```

```python
interp_func = interp1d(offset_to_match, mean_intensity, kind='cubic')
```

```python
fine_offsets = np.linspace(min(offset_to_match), max(offset_to_match), num=2000)
smoothed_intensity = interp_func(fine_offsets)
```

```python
print(max(smoothed_intensity), min(smoothed_intensity))
length_val = max(smoothed_intensity) - min(smoothed_intensity)
length_val_half = length_val/2
half_max = max(smoothed_intensity) - length_val_half
half_max
```

```python
difference = []
index = []
for i, intensity in enumerate(smoothed_intensity):
    diff = half_max - intensity
    if ((diff >= 0) & (diff <=1)):
        difference.append(diff)
        index.append(i)
    
print(difference)
print(index)
```

```python
spline_rep = splrep(offset_to_match_corr, mean_intensity, k=3)
fine_offsets = np.linspace(min(offset_to_match_corr), max(offset_to_match_corr), num=1000)
smoothed_intensity = splev(fine_offsets, spline_rep)
```

```python
max_intensity = max(smoothed_intensity)
min_intensity = min(smoothed_intensity)
half_max = (max_intensity + min_intensity) / 2
print(min_intensity)
print(max_intensity - min_intensity)
```

```python
crossings = np.where(np.diff(np.sign(smoothed_intensity - half_max)))[0]
```

```python
x_crossings = []
for crossing in crossings:
    x1, x2 = fine_offsets[crossing], fine_offsets[crossing + 1]
    y1, y2 = smoothed_intensity[crossing], smoothed_intensity[crossing + 1]
    slope = (y2 - y1) / (x2 - x1)
    x_cross = x1 + (half_max - y1) / slope
    x_crossings.append(x_cross)
```

```python
# bessel function
def func(theta):
    x = (np.pi * np.sin(np.radians(theta)))/ (0.21/2.3)
    return ((2*scipy.special.j1(x))/x)**2
```

```python
theta=np.linspace(-24,24,300)
p = func(theta)
plt.figure(figsize = (10,7))
plt.plot(theta, (p*458.7) + 556.5)
plt.title("Theoretical beam pattern expected for a uniformly illuminated circular aperture")
plt.xlabel("Angular offsets (in degrees)")
plt.ylabel("Theoretical average integrated intensity")
plt.show()
```

```python
#Calculate bin width for theoretical
y_values_theoretical = (p*458.7) + 556.5
x_values_theoretical = theta
max_y_value_theoretical = (np.max(y_values_theoretical)-np.min(y_values_theoretical)) / 2+np.min(y_values_theoretical)

p_theoretical = np.where(y_values_theoretical>=max_y_value_theoretical)[0]
bin_width_theoretical = x_values_theoretical[p_theoretical[0]]-x_values_theoretical[p_theoretical[-1]]
print(f'Theoretical half-power beam width: {bin_width_theoretical}')
```
